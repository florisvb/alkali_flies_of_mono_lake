```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

r = 0.5 # um
l = 2 # um
ca = np.linspace(0,180,50) # deg
st = 0.0728 # surface tension, J / m^2

ca *= np.pi/180.

phi = -1*(np.pi-np.arcsin(2*r*np.sin(ca)/l) - ca)

fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(ca*180/np.pi, phi*180/np.pi)
```
